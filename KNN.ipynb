```python
data.describe()
```

```python
# 이 세가지 방법 중에서 가장 일반적인 방법은 dropna()를 이용한 결측치 행 제거
# 평균등을 이용해 지우는 3번째 방법은 실젯값과 일치할 가능성이 매우 낮아 오차의 원인이 된다.(noise가 더해진 효과)
# 변수자체를 drop()하는 방법은 매우 무모함. 

# 만약 특정 변수의 99%가 결측치라면? -> 이럴 때는 변수 자체를 제거하는 방식이 합리적.
# 하지만 경우에 따라서 결측치가 90%이상이더라도 해당 변수가 매우 중요한 역할이라고 생각이 들면 어떻게든 활용 방법을 찾는 것이 좋음.

# 본 예제에서는 무난한 방법인, fillna()를 이용하여 중앙값을 채우는 방식으로 할 것임.
data.fillna(data.median(), inplace=True)
```

Data Scailing 데이터 스케일링

```python
data.describe()
```

```python
# Scailing을 할 때에는 Scikit learn 라이브러리의 preprocessing 모듈을 사용할 예정.
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
# StandardScaler: 평균이 0이되고, 표준편차가 1이 되도록 데이터를 고르게 분포시키는 데 사용
# MinMaxScaler: 데이터 분포의 특성을 최대한 그대로 유지하고 싶을 때 사용
# RobustScaler: 데이터에 아웃라이어가 존재하고, 그 영향력을 그대로 유지하고 싶을 때 사용
```

```python
st_scaler = StandardScaler()
# 각 column별 평균, 표준편차로 데이터 학습
st_scaler.fit(data)
# 이제 data를 변환함(transform)
st_scaled = st_scaler.transform(data)
```

```python
st_scaled
```

```python
pd.DataFrame(st_scaled)
```

```python
st_scaled = pd.DataFrame(st_scaled, columns= data.columns)
st_scaled
```

```python
round(st_scaled.describe(), 2)
```

```python
# Robust Scaling: outlier에 견고하게 scaling
rb_scaler = RobustScaler()
rb_scaled = rb_scaler.fit_transform(data)
rb_scaled = pd.DataFrame(rb_scaled, columns= data.columns)
```

```python
rb_scaled
```

```python
# scaling시 평균대신, 중앙값을 사용하여 outlier에 더 robust하게 scaling이 가능
round(rb_scaled.describe(), 2)
```

```python
#minMax scaler는 최대값이 1.0 최솟값이 0.0으로 fitting됨!
mM_scaler = MinMaxScaler()
mM_scaled = mM_scaler.fit_transform(data)
mM_scaled = pd.DataFrame(mM_scaled, columns= data.columns)
round(mM_scaled.describe(), 2)
```

아웃라이어의 영향이 큰 데이터고, 이를 피하고 싶으면 Robust Scaling <br><br>
Standard Scaling은 평균과 표준편차, <br><br>
minMax Scaling은 최대값/최소값, <br><br>
Robust Scaling은 사분위값을 사용하기 때문에 outlier에 영향을 덜 받는다.<br><br>

* minMax Scaling은 기존의 데이터 분포의 특징을 최대한 가져가고 싶을떄! <br><br>
  standard 방식은 0을 기준으로 좌우대칭형이기 때문에 outlier의 영향을 받으면 특징 상실!

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis=1), data['class'], test_size=0.2, random_state=100)
```

```python
mM_scaler = MinMaxScaler()
mM_scaler.fit(X_train)

X_train_scaled = mM_scaler.fit_transform(X_train)
X_test_scaled = mM_scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
score = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    A_score = accuracy_score(y_test, pred)
    score.append(A_score)
    print(f"i를 {i}로 설정하였을 때, 정확도는 {A_score}")
```

```python
score
```

```python
sns.lineplot(x=range(1, 21), y=score)
```
