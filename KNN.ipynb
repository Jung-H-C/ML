```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis=1), data['class'], test_size=0.2, random_state=100)
```

```python
mM_scaler = MinMaxScaler()
mM_scaler.fit(X_train)

X_train_scaled = mM_scaler.fit_transform(X_train)
X_test_scaled = mM_scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
score = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    A_score = accuracy_score(y_test, pred)
    score.append(A_score)
    print(f"i {i}  ,  {A_score}")
```

```python
score
```

```python
sns.lineplot(x=range(1, 21), y=score)
```
