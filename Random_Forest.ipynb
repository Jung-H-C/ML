```python
def isFloat(value):
    try:
        num = float(value)
        return num
    # 에러가 발생시 except 구문을 실행합니다. (if절이랑 비슷)
    except ValueError:
        return np.NaN
```

```python
def mile(x):
    if x['fuel'] == 'Petrol':
        return x['mileage'] / 80.43
    elif x['fuel'] == 'Diesel':
        return x['mileage'] / 73.56
    elif x['fuel'] == 'LPG':
        return x['mileage'] / 40.85
    else:
        return x['mileage'] / 44.23 # CNG
```

```python
data['mileage'] = data.apply(mile, axis=1)
```

```python
# mileage_unit은 필요없으니 지운다 (달러당 주행거리로 통일시킴)
data.drop('mileage_unit', axis=1, inplace=True)
```

```python
def torque_unit(x):
    if 'NM' in str(x):
        return 'Nm'
    elif 'KGM' in str(x):
        return 'kgm'
```

```python
data['torque_unit'] = data['torque'].apply(torque_unit)
```

```python
# 문자열에 숫자, char 같이 들어간 text 분리하는 함수
def split_num(x):
    x = str(x)
    # x의 i번째 index 값 j
    for i, j in enumerate(x):
        if j not in '0123456789':
            cut = i
            break
    return x[:cut]
```

```python
data['torque'] = data['torque'].apply(split_num)
```

```python
data['torque']
```

```python
data['torque'] = data['torque'].astype('float64')
```

```python
data['torque'] = data['torque'].replace('', np.NaN)
```

```python
data['torque'] = data['torque'].astype('float64')
```

```python
data['torque'].head()
```

```python
# torque 계산 함수
def torque_trans(x):
    if x['torque_unit'] == 'kgm':
        return x['torque'] * 9.8066
    else:
        return x['torque'] 
```

```python
data['torque'] = data.apply(torque_trans, axis=1)
```

```python
def split_brand(x):
    x = str(x)
    for i, j in enumerate(x):
        if j == ' ':
            cut = i
            break
    return x[:cut]
```

```python
data['brand'] = data['name'].apply(split_brand)
data.head()
```

```python
# 브랜드가 target variable에 미치는 영향이 클 것으로 보아 평균값으로는 결측치를 채우기 보다는 과감히 행 삭제!
data.dropna(inplace=True)
len(data)
```

```python
data = pd.get_dummies(data, columns=['brand', 'fuel', 'seller_type', 'owner', 'transmission'], drop_first=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('selling_price', axis=1), data['selling_price'], test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=90)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print("train_rmse:", mean_squared_error(y_train, train_pred) ** 0.5, "test_rmse", mean_squared_error
      (y_test, test_pred) ** 0.5)
```
